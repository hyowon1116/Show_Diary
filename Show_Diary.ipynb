#### 0) set_part() 함수: 마지막 글자의 받침 유무에 따라 조사 '을', '를'을 결정하는 함수

```python
def set_part(word: str):
    consonant = (ord(word[-1]) - 44032) % 28
    if consonant:
        return '을'
    return '를'
```

#### 1) create_show() 함수: 관람했거나 관람할 공연을 데이터에 등록하는 기능

```python
def create_show():
    print('공연 추가를 선택하셨습니다.')
    
    # 관람 여부 선택
    # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
    while True:
        print('\n< 관람여부 >')
        print('1. 이미 관람한 공연\n2. 나중에 관람할 공연\n0. 뒤로 가기')
        answer1 = input('=> 입력: ')
        if answer1 in ['1', '2', '0']:
            break
        print('보기에 없는 번호입니다. 다시 확인해주세요.')
    
    # 뒤로 가기 -> 함수 종료
    if answer1 == '0':
        return
    
    global show
    
    # 공연명, 공연지역명, 공연일자, 예매금액, 장르명 입력
    print('\n< 공연명 >')
    answer2 = input('=> 입력: ')
    print('\n< 공연지역명 >')
    answer3 = input('=> 입력: ')
    print('\n< 공연일자 > (ex. 2022-01-01)')
    answer4 = input('=> 입력: ')
    print('\n< 예매금액 >')
    answer5 = int(input('=> 입력: '))
    print('\n< 장르명 >')
    answer6 = input('=> 입력: ')
    
    # 이미 관람한 공연 -> 공연별점 입력, 관람여부 1
    # 나중에 관람할 공연 -> 공연별점 NaN, 관람여부 0
    if answer1 == '1':
        print('\n< 공연별점 > (5점 만점)')
        answer7 = float(input('=> 입력: '))
        watch = 1
    else:
        answer7 = np.NaN
        watch = 0
    
    # 기존 데이터에 공연 추가
    show = show.append({
        '관람여부': watch,
        '공연명': answer2,
        '공연지역명': answer3,
        '공연일자': answer4,
        '예매금액': answer5,
        '장르명': answer6,
        '공연별점': answer7
    }, ignore_index = True)
    
    print(f"\n공연 '{answer2}'{set_part(answer2)} 추가했습니다!")
```

#### 2) read_show() 함수: 공연을 조회해 목록을 출력하는 기능

```python
def read_show():
    print('공연 조회를 선택하셨습니다.')
    
    # 출력 기능 선택
    # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
    while True:
        print('\n원하는 출력 기능을 선택해주세요.')
        print('1. 전체 출력\n2. 조건별 출력\n0. 뒤로 가기')
        answer = input('=> 입력: ')
        if answer in ['1', '2', '0']:
            break
        print('보기에 없는 번호입니다. 다시 확인해주세요.')
    
    # 뒤로 가기 -> 함수 종료
    if answer == '0':
        return
    
    global show
    
    # 전체 출력 -> print_show() 함수 실행
    # 조건별 출력 -> 조건 기준 항목 선택
    if answer == '1':
        print_show()
    else:
        numList = ['공연지역명', '예매금액', '장르명', '공연별점', '관람여부']
        print('조건별 출력을 선택하셨습니다.\n')
        
        # 조건 기준 선택
        # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
        while True:
            print('조건 기준을 선택해주세요.')
            for i in range(5):
                print(f'{i + 1}. {numList[i]}')
            number = int(input('=> 입력: '))
            if number in range(1, 6):
                break
            print('보기에 없는 번호입니다. 다시 확인해주세요.\n')
        
        # 공연지역명, 장르명 -> 해당 항목과 입력한 값이 일치하는 공연 출력
        # 예매금액, 공연별점 -> 해당 항목이 하한값 이상 상한값 이하인 공연 출력
        # 관람여부 -> 관람한 공연 또는 관람할 공연 출력
        if number in [1, 3]:
            name = input(f'=> {numList[number - 1]} 입력: ')
            show_data = show[show[numList[number - 1]] == name].reset_index()
            show_data = show_data[['관람여부', '공연명', '공연지역명', '공연일자', '예매금액', '장르명', '공연별점']]
            print(f"\n{numList[number - 1]}이 '{name}'인 공연 목록입니다.")
        elif number in [2, 4]:
            bottom = float(input(f'=> {numList[number - 1]} 하한 입력: '))
            top = float(input(f'=> {numList[number - 1]} 상한 입력: '))
            show2 = show[show[numList[number - 1]] <= top]
            show3 = show2[bottom <= show2[numList[number - 1]]].reset_index()
            show_data = show3[['관람여부', '공연명', '공연지역명', '공연일자', '예매금액', '장르명', '공연별점']]
            if number == 2:
                unit = '원'
            else:
                unit = '점'
            print('\n%s이 %g%s 이상 %g%s 이하인 공연 목록입니다.'%(numList[number - 1], bottom, unit, top, unit))
        else:
            # 관람 여부 선택
            # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
            while True:
                print('\n관람 여부를 선택해주세요.')
                print('1. 이미 관람한 공연\n2. 나중에 관람할 공연')
                watch = input('=> 입력: ')
                if watch in ['1', '2']:
                    break
                print('보기에 없는 번호입니다. 다시 확인해주세요.')
            
            # 이미 관람한 공연 -> 관람여부 항목이 1인 공연 출력 (관람여부 항목 생략)
            # 나중에 관람할 공연 -> 관람여부 항목이 0인 공연 출력 (관람여부, 공연별점 항목 생략)
            if watch == '1':
                print('\n관람한 공연 목록입니다.')
                show2 = show.loc[show['관람여부'] == 1].reset_index()
                show_data = show2[['공연명', '공연지역명', '공연일자', '예매금액', '장르명', '공연별점']]
            else:
                print('\n관람 예정인 공연 목록입니다.')
                show2 = show.loc[show['관람여부'] == 0].reset_index()
                show_data = show2[['공연명', '공연지역명', '공연일자', '예매금액', '장르명']]
        
        display(show_data)
```

#### 3) update_show() 함수: 공연 데이터에서 변화된 항목을 수정하는 기능

```python
def update_show():
    print('공연 수정을 선택하셨습니다.')
    
    # 수정 기능 선택
    # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
    while True:
        print('\n원하는 수정 기능을 선택해주세요.')
        print('1. 공연 항목 변경\n2. 공연 관람 여부 변경\n0. 뒤로 가기')
        answer = input('=> 입력: ')
        if answer in ['1', '2', '0']:
            break
        print('보기에 없는 번호입니다. 다시 확인해주세요.')
    
    # 뒤로 가기 -> 함수 종료
    if answer == '0':
        return
    
    # 공연 항목 변경 -> 특정 공연 선택
    # 공연 관람 여부 변경 -> move_show() 함수 실행
    if answer == '1':
        global show
        
        print('\n전체 공연 목록입니다. (관람여부: 1은 관람한 경우, 0은 관람 예정인 경우)')
        display(show)
        
        # 변경할 공연 선택
        # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
        while True:
            print(f'변경할 공연의 번호를 입력해주세요. (0 ~ {len(show) - 1})')
            number1 = int(input('=> 입력: '))
            if number1 in range(len(show)):
                break
            print('보기에 없는 번호입니다. 다시 확인해주세요.\n')
        
        name = show.loc[number1]['공연명']
        print(f"공연 '{name}'{set_part(name)} 선택하셨습니다.")
        
        # 변경할 공연이 관람한 공연인 경우 -> 관람여부 외 항목에서 선택 가능
        # 변경할 공연이 관람할 공연인 경우 -> 관람여부, 공연별점 외 항목에서 선택 가능
        if show.loc[number1]['관람여부']:
            text = ['공연명', '공연지역명', '공연일자', '예매금액', '장르명', '공연별점']
        else:
            text = ['공연명', '공연지역명', '공연일자', '예매금액', '장르명']
        display(show.loc[number1][text])
        
        # 변경할 항목 선택
        # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
        while True:
            print('변경할 항목의 번호를 입력해주세요.')
            for i in range(len(text)):
                print(f'{i + 1}. {text[i]}')
            number2 = int(input('=> 입력: '))
            if number2 in range(1, len(text) + 1):
                break
            print('보기에 없는 번호입니다. 다시 확인해주세요.\n')
        
        print('\n변경할 내용을 입력하세요.')
        change = input('=> 입력: ')
        
        # 변경할 항목이 예매금액 또는 공연별점일 경우 -> 입력값을 숫자형으로 변경
        if answer in [4, 6]:
            change = float(change)
        
        # 기존 데이터에 변경 항목 대입
        show.iloc[number1, number2] = change
        print('\n변경을 완료했습니다!')
    else:
        move_show()
```

#### 4) delete_show() 함수: 특정 공연을 선택해 삭제하는 기능

```python
def delete_show():
    print('공연 삭제를 선택하셨습니다.')
    
    # 삭제 기능 선택
    # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
    while True:
        print('\n원하는 삭제 기능을 선택해주세요.')
        print("1. 조건별 삭제\n2. 관람 예정 일괄 삭제\n0. 뒤로 가기")
        answer = input('=> 입력: ')
        if answer in ['1', '2', '0']:
            break
        print('보기에 없는 번호입니다. 다시 확인해주세요.')
    
    # 뒤로 가기 -> 함수 종료
    if answer == '0':
        return
    
    # 조건별 삭제 -> 특정 공연 선택
    # 관람 예정 일괄 삭제 -> erase_show() 함수 실행
    if answer == '1':
        global show
        
        print('\n기존 공연 목록입니다.')
        display(show)
        
        # 기존 데이터에 공연이 존재할 경우
        if len(show):
            
            # 삭제할 공연 선택
            # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
            while True:
                print(f'삭제하실 공연의 번호를 입력해주세요. (0 ~ {len(show) - 1})\n(여러 개 입력할 시 번호마다 띄어쓰기 필수)')
                numbers = input('=> 입력: ').split()
                check = True
                
                for n in numbers:
                    if int(n) not in range(len(show)):
                        print('보기에 없는 번호가 포함되어 있습니다. 다시 확인해주세요.\n')
                        check = False
                        break
                
                if check:
                    numbers = map(int, numbers)
                    break
            
            # 삭제할 공연의 공연명을 리스트에 저장
            # 데이터에서 공연 삭제
            title = []
            for n in numbers:
                title.append(show.loc[n]['공연명'])
                show = show.drop(n)
            
            show = show.reset_index()[['관람여부', '공연명', '공연지역명', '공연일자', '예매금액', '장르명', '공연별점']]
            print("\n공연 '" + "', '".join(title) + "'" + set_part(title[-1]) + " 삭제했습니다!")
        
        # 기존 데이터에 공연이 존재하지 않을 경우
        else:
            print('삭제할 수 있는 공연이 없습니다.')
    else:
        erase_show()
```

#### 5) print_show() 함수: 정렬 기준을 선택해 전체 공연 목록을 출력하는 기능

```python
def print_show():
    print('전체 출력을 선택하셨습니다.')
    numList = ['생성순서', '공연명', '공연지역명', '공연일자', '예매금액', '장르명', '공연별점']
    
    # 정렬 기준 선택
    # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
    while True:
        print('\n정렬 기준을 선택해주세요.')
        for i in range(7):
            print(f'{i + 1}. {numList[i]}')
        answer = input('=> 입력: ')
        if answer in ['1', '2', '3', '4', '5', '6', '7']:
            break
        print('보기에 없는 번호입니다. 다시 확인해주세요.')
    
    global show
    arr = numList[int(answer) - 1]
    print(f"\n'{arr}'{set_part(arr)} 기준으로 정렬한 공연 목록입니다.\n(관람여부: 1은 관람한 경우, 0은 관람 예정인 경우)")
    
    # 생성순서 -> 기존 데이터 출력
    # 그 외 -> 해당 항목을 기준으로 정렬한 데이터 출력
    if answer == '1':
        display(show)
    else:
        display(show.sort_values(arr))
```

#### 6) move_show() 함수: 관람 예정 공연을 관람한 공연으로 옮기는 기능

```python
def move_show():
    print('공연 관람 여부 변경을 선택하셨습니다.')
    print('\n관람 예정 공연 목록입니다.')
    
    # 관람여부 항목이 0인 공연 출력
    global show
    show2 = show.loc[show['관람여부'] == 0].reset_index()
    display(show2[['공연명', '공연지역명', '공연일자', '예매금액', '장르명']])
    
    if len(show2):
        # 관람한 공연 선택
        # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
        while True:
            print(f'관람하신 공연의 번호를 입력해주세요. (0 ~ {len(show2) - 1})')
            number = int(input('=> 입력: '))
            if number in range(len(show2)):
                break
            print('보기에 없는 번호입니다. 다시 확인해주세요.\n')
        
        print(f"공연 '{show2.loc[number]['공연명']}'을(를) 선택하셨습니다.")
        
        # 공연별점 입력
        print('\n< 공연별점 > (5점 만점)')
        star = float(input('=> 입력: '))
        
        # 기존 데이터에 변경 항목 대입
        idx = show2.iloc[number, 0]
        show.iloc[idx, 0] = 1
        show.iloc[idx, 6] = star
        
        name = show2.loc[number]['공연명']
        print(f"\n공연 '{name}'{set_part(name)} 관람 예정 목록에서 관람 목록으로 옮겼습니다!")
    else:
        print('관람 예정 공연이 없습니다.')
```

#### 7) erase_show() 함수: 과거에 관람 예정이었던 공연을 일괄적으로 삭제하는 기능

```python
def erase_show():
    
    global show
    num = []
    
    # 관람여부가 0인 공연 중 공연일자가 과거인 공연의 번호를 리스트에 저장
    for i in show.index:
        if show.loc[i]['관람여부'] == 1:
            continue
        date = show['공연일자'][i]
        y, m, d = int(date[:4]), int(date[5:7]), int(date[8:])
        if datetime.now() > datetime(y, m, d):
            num.append(i)
    
    # 데이터에서 공연 삭제
    for j in num:
        show = show.drop(j)
    
    print('\n과거에 관람 예정인 공연을 일괄 삭제했습니다!')
    show = show.reset_index()[['관람여부', '공연명', '공연지역명', '공연일자', '예매금액', '장르명', '공연별점']]
```

#### 8) start_show() 함수: Show Diary 프로그램을 시작하며 데이터를 불러오거나 생성하는 기능

```python
def start_show():
    
    # 데이터 불러오기 여부 선택
    # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
    while True:
        print('이전의 데이터를 불러오시겠습니까?')
        print('1. 네\n2. 아니요')
        answer = input('=> 입력: ')
        if answer in ['1', '2']:
            break
        print('보기에 없는 번호입니다. 다시 확인해주세요.\n')
    
    if answer == '1':
        # 데이터 이름 입력
        # 존재하지 않는 데이터를 입력하면 다시 입력하도록 반복문 설정
        while True:
            print('\n불러올 데이터의 이름을 작성해주세요. (단, csv 형식만 가능)')
            name = input('=> 입력: ')
            try:
                show_data = pd.read_csv(f'{name}.csv')
                print(f"기존에 있던 '{name}.csv'를 불러왔습니다.")
                break
            except FileNotFoundError:
                print(f"'{name}.csv' 파일은 존재하지 않습니다. 다시 확인해주세요.")
    else:
        # 새로운 데이터 생성
        show_data = pd.DataFrame({
            '관람여부': [],
            '공연명': [],
            '공연지역명': [],
            '공연일자': [],
            '예매금액': [],
            '장르명': [],
            '공연별점': []
        })
        print('\n새로운 데이터를 생성했습니다.')
    
    return show_data
```

#### 9) exit_show() 함수: 데이터 저장 여부를 물어보고 Show Diary 프로그램을 종료하는 기능

```python
def exit_show():
    # 저장 여부 선택
    # 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
    while True:
        print('\n지금까지의 공연 데이터를 저장하시겠습니까?')
        print('1. 저장한다\n2. 저장하지 않는다')
        answer = input('=> 입력: ')
        if answer in ['1', '2']:
            break
        print('보기에 없는 번호입니다. 다시 확인해주세요.\n')
    
    # 저장한다 -> 데이터 이름 입력 및 데이터 추출 후 프로그램 종료
    # 저장하지 않는다 -> 프로그램 종료
    if answer == '1':
        global show
        print('\n추출할 공연 데이터의 이름을 작성해주세요.')
        name = input('=> 입력: ')
        show.to_csv(f'{name}.csv', sep = ',', encoding = 'utf-8-sig', index = False)
        print(f"지금까지의 데이터를 '{name}.csv'로 추출했습니다.")
    
    print('\n<Show Diary>를 종료합니다. 이용해주셔서 감사합니다.')
```

## 3. 프로그램 실행

```python
print('< Show Diary >')
print('공연 데이터를 관리하는 Show Diary 프로그램을 실행합니다.\n')

show = start_show()

# 기능 선택
# 주어진 보기 외 다른 답변을 하면 다시 입력하도록 반복문 설정
while True:
    print('\n원하는 기능을 선택하세요.')
    print('1. 공연 추가')
    print('2. 공연 수정')
    print('3. 공연 삭제')
    print('4. 공연 조회')
    print('0. 프로그램 종료')
    
    choice = input('=> 입력: ')
    if choice == '1':
        create_show()
    elif choice == '2':
        update_show()
    elif choice == '3':
        delete_show()
    elif choice == '4':
        read_show()
    elif choice == '0':
        break
    else:
        print('보기에 없는 번호입니다. 다시 확인해주세요.')

exit_show()
```
